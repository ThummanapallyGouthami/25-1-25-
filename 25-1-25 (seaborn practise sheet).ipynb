```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
print(tips.head())
```

```python
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color="skyblue")
plt.title("Distribution of total bills")
plt.show()
```

```python
"""sns.kdeplot(data=tips,x="total_bill",shade=True,color='lightgreen')
plt.title("kde plot of total bills")
plt.show()"""


sns.kdeplot(data=tips,x="total_bill",fill=True,color='lightgreen')
plt.title("kde plot of total bills")
plt.show()
```

```python
sns.barplot(data=tips,x="day",y="total_bill",palette="viridis")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette="coolwarm")
plt.title("Box plot of total Bills by day and Gender")
plt.show()
```

```python
# scatter plot
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",palette="deep")
plt.title("scatter plot of total Bill vs Tip")
plt.show()
```

```python
#line plot
sns.lineplot(data=tips,x="size",y="tip",marker='o',color='green')
plt.title("Line plot of party size vs .tip")
plt.show()
```

```python
#Pair plot
sns.pairplot(data=tips,hue="sex",palette="husl")
plt.show()
```

```python
#Heat map
numeric_tips = tips.select_dtypes(include=["number"])
correlation_matrix = numeric_tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("train.csv")
pdf.head()
```

```python
sns.barplot(data=df,x='Pclass',y='Survived',errorbar=None,palette='coolwarm')
#sns.barplot(data=df, x='Pclass', y='Survived', color='green') 
plt.title("survival rate by passenger class")
plt.xlabel('passenger class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Fare', color='lightblue', label='Age')
sns.scatterplot(data=df, x='Fare', y='Age', color='lightgreen', label='Fare')
plt.title("Scatter plot between Age and Fare")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Fare',hue='Survived',palette='coolwarm',alpha=0.6)
plt.title("Age vs. Fare")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass']], hue='Pclass',palette='coolwarm',diag_kind='kde')
plt.suptitle('pairwise relationships',y=1.02)
plt.show()
```

```python
numeric = df.select_dtypes(include=[int,float])
correlation_matrix = numeric.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

```
